In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:

Note that you'll need to clean and format the new data.

Observation:

- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
import pandas as pd

# Load the dataset
url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
df_1 = pd.read_csv(url)


url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
df_2 = pd.read_csv(url)


url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'
df_3 = pd.read_csv(url)
```

```python
df_1.shape
```

```python
df_2.shape
```

```python
df_3.shape
```

```python
final_df = pd.concat([df_1, df_2, df_3], axis=0)
final_df.shape
```

```python
def fix_col_name(final_df):
    final_df.columns = final_df.columns.str.strip().str.lower().str.replace(' ', '_')
    final_df.rename(columns= {'st': 'state'}, inplace=True)
    return final_df

final_df = fix_col_name(final_df)
final_df.columns
```

```python
final_df.dtypes
```

```python
final_df['gender'].unique()
```

```python
final_df['gender'].str.replace('Femal', 'F').str.replace('Male', 'M').str.replace('Female', 'F').str.replace('Male', 'M')
```

```python
final_df['customer_lifetime_value'] = final_df['customer_lifetime_value'].str.replace('%','') 
final_df
```

```python
final_df['customer_lifetime_value'] = final_df['customer_lifetime_value'].astype(float)
final_df['number_of_open_complaints'] = final_df['number_of_open_complaints'].apply(lambda x: x.split("/")[1] if isinstance(x,str) and len(x.split("/")) > 1 else x)
final_df['number_of_open_complaints'] = final_df['number_of_open_complaints'].astype(float)
```

```python
final_df.dtypes
```

```python
final_df.isnull() 
final_df.dropna()
final_df.shape
```

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
# Load the dataset
url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv'
df = pd.read_csv(url)

#First look at dataset

df.head(10)
```

```python
#Check for datatypes
df.dtypes
```

```python
#unnamed:_0 column seems useless, we can consider it is safe to drop it
del df["unnamed:_0"]
```

```python
df['customer_lifetime_value'] = df['customer_lifetime_value'].round(2)
df['customer_lifetime_value']
```

```python
df['total_revenue'] = df['months_since_policy_inception'] * df['monthly_premium_auto']
df.head(10)
```

```python
print(df.columns)
```

```python
pivot_df = df.pivot_table(index='sales_channel', columns='vehicle_type', values=['total_revenue'], aggfunc='sum')
pivot_df
```

```python
df.groupby('sales_channel')['total_revenue'].agg('sum')
df.groupby
```

```python
#A pivot table that shows the average customer lifetime value per gender and education level


pivot_df = df.pivot_table(index='education', columns='gender', values=['customer_lifetime_value'], aggfunc='mean')
pivot_df

#CLV increases corresponding to the level of education, there is no significant difference in CLV between genders except of College and HighSchool (we need additional data to make deeper conclusions about gender inequality)
```
